Non-Homothetic consumption model
Resource constraints, $Y_i$ produced by competitive firm equals the sum of all the $C_i^h$ demand across household $h$
$$Y_i=\sum^{\mathcal{H}}_{h=1} C_i^h$$

### Households
Households maximize their implicit utility with respect to their level of expenditures. Higher expenditure mean higher revenue and higher utility. 

Type $h \in \mathcal{H}$ household maximizes utility given budget constraint

$$ \max\limits_{\{C_1^h\}^\mathcal{I}_{i=1}} U(C_1^h,...,C_\mathcal{I}^h) $$
$$ st \quad E^h \geq \sum^{\mathcal{I}}_{i=1} p_iC_i^h.$$

In this model, household differs by their productivity level. We evaluate household productivity level for an individual by looking at their individual production. Suppose $y^h$ correspond to individual $h$ production then $\frac{y^h}{\theta^h}=L^h$ accounts for number of hour worked with $\theta^h$ the individual productivity type. higher $\theta^h$ accounts for higher productivity and lower hours worked $L^h$ for a same production $y^h$. Let us further assume that a distribution and thus a ranking is available so that and higher indexed  $h \in \mathcal{H}$ constitutes a higher value for household expenditure level ($E^h$). 
To account for an income effect in the allocation of resources across consumption goods, I adopt the implicit utility function with non-homothetic parameter proposed by  \cite{sato_homothetic_1977}\cite{comin_structural_2021}:
$$\sum^{\mathcal{I}}_i \gamma_i^{\frac{1}{\sigma}} \left[ \frac{C_i^h}{g(U)^{\epsilon_i}}\right]^{\frac{\sigma-1}{\sigma}}=1.$$
Let $U=U(C_1^h,...,C^h_\mathcal{I})$ The maximization problem would take the following form

$$\mathcal{L} = U+\rho[1-\sum^{\mathcal{I}}_i \gamma_i^{\frac{1}{\sigma}} \left[ \frac{C_i}{g(U)^{\epsilon_i}}\right]^{\frac{\sigma-1}{\sigma}}] + \lambda [E^h-\sum^{\mathcal{I}}_{i}p_iC_i^h ].$$
 When we derive with respect to $C_i^h$ it gives

$$\frac{\rho}{\lambda} \frac{1-\sigma}{\sigma}\omega_i^h=p_iC_i^h$$
where $\omega_i^h=\gamma_i^{\frac{1}{\sigma}} \left[ \frac{C_i^h}{g(U)^{\epsilon_i}}\right]^{\frac{\sigma-1}{\sigma}}$ and stands for share of consumption in good $i$ by household type $h$ Since $\sum^{\mathcal{I}}_{i} \omega_i=1$ then we get 
$$E^h\equiv \sum^{\mathcal{I}}_{i} p_ic_i=\frac{\rho}{\lambda} \frac{1-\sigma}{\sigma}.$$

Then $E^h\omega_i^h=p_iC_i^h$ so that  $E^h\gamma_i^{\frac{1}{\sigma}} \left[ \frac{C_i^h}{g(U)^{\epsilon_i}}\right]^{\frac{\sigma-1}{\sigma}}=p_iC_i^h$ and isolating $C_i$ gives the Hicksian demand function
$$C_i^h=\left(\frac{p_i}{E^h}\right)^{-\sigma} \gamma_i g(U)^{\epsilon_i(1-\sigma)}.$$

 Substituting this demand into the expenditure share, we obtain the true expenditure share
$$\omega_i^h=\gamma_i\left[\frac{p_i}{E^h}g(U)^{\epsilon_i}\right]^{(1-\sigma)}.$$

 Total expenditure by household $h$ is then 
$$E^h \equiv \sum^{\mathcal{I}}_{i=1}p_iC_i^h= \left[\sum^{\mathcal{I}}_{i=1}\gamma_i\left(p_ig(U)^{\epsilon_i}\right)^{1-\sigma}\right]^{\frac{1}{1-\sigma}}$$

Relative demand $\forall i,j \in \mathcal{I}$ for household $h$ is then found by taking the log of $\frac{C^h_i}{C^h_j}$ and delivers us: 
$$\log\left(\frac{C^h_i}{C^h_j}\right)= \sigma\log\left(\frac{p_j}{p_i}\right) + (\epsilon_i-\epsilon_j)(1-\sigma)\log g(U) + \log\left(\frac{\gamma_i}{\gamma_j}\right).$$

 We consequently have the elasticity of relative demand with respect to a monotonic transformation of $g(.)$ equals
$$\frac{\partial \log\left(\frac{C^h_i}{C^h_j}\right)}{\partial \log g(U)}=(1-\sigma)(\epsilon_i-\epsilon_j).$$
We can also see that the elasticity of relative demand with respect to change in relative prices equals 
$$\frac{\partial \log\left(\frac{C^h_i}{C^h_j}\right)}{\partial \log \left(\frac{p_j}{p_i}\right)}=\sigma.$$

 Relative Hicksian demand on expenditure shares yield
$$\log \left(\frac{\omega^h_i}{\omega^h_j}\right)=\log \left(\frac{\gamma_i}{\gamma_j}\right)+(1-\sigma) \log \left(\frac{p_i}{p_j}\right) + (1-\sigma)(\epsilon_i-\epsilon_j) \log g(U)$$

 Let us take a good as reference good indexed as $b \in \mathcal{I}$. We could then evaluate the relative demand of other goods with respect to the reference one as goods $\forall i \in \mathcal{I}_{-b}$

$$\log \omega^h_i=\log \gamma_i +(1-\sigma) \log \left(\frac{p_i}{E^h}\right) +(1-\sigma)\epsilon_i \log g(U) , \quad \forall i \in \mathcal{I}. $$

by taking a reference good $b$ we find:
$$\log \omega^h_i=\frac{\epsilon_i}{\epsilon_b} \log \omega_b^h +(1-\sigma)\left(\frac{\epsilon_i}{\epsilon_b}-1\right)\log\left(\frac{E^h}{p_b}\right)+\log \left(\frac{\gamma_i}{\gamma_b^{\frac{\epsilon_i}{\epsilon_b}}}\right)$$

```julia
struct Calibration
    σ           #elasticity of relative demand with respect to price
    ϵ_p         #elasticity of relative demand with respect to income in primary necessities sector
    ϵ_n         #elasticity of relative demand with respect to income in normal good sector
    ϵ_l         #elasticity of relative demand with respect to income in luxury good sector
    A_p         #productivity in primary necessities sector
    A_n         #productivity in primary necessities sector
    A_l         #productivity in primary necessities sector
    shock_size  #random shock on price size in term of standard deviation 
    g_primary   #inflation primary goods
    g_normal    #inflation normal goods
    g_luxury    #inflation luxury goods
end
```

```julia
cal = Calibration(
    0.5,        # sigma=0.5 completementary goods
    0.7,        # demand share for necessities decrease with increasing utility 
    1.0,        # Constant return, constant demand share with increasing utility
    1.25,        # Increasing return, higher demand share with increasing utility
    1.0,        # lower productivity level for necessisties
    1.2,
    1.4,
    5,
    1.05,
    1.02,
    1.01
)
```

### Price dynamics

Let us adopt 3 goods with different price dynamics so that we can look for the impact of relative prices on household consumption behavior.

```julia
#Price dynamics
using Plots
using DataFrames

#Time and number of household considered
T=20
N=3

#Utility differs by their level of revenue indexed by i 
u=[]
for i=1:N
    utility= log(1+i^10)
    u=push!(u, utility)
end

#Get our price level dynamics
index=100.0 
pp=[index]  #price primary
pn=[index]  #price normal
pl=[index]  #price luxury
agg=[index] #aggregate price with homothetic
for t=1:T-1
    p_p=100*cal.g_primary^(t)+cal.shock_size*randn() # price dynamics of primary goods
    p_n=100*cal.g_normal^(t)+cal.shock_size*randn() # price dynamics of normal goods
    p_l=100*cal.g_luxury^(t)+cal.shock_size*randn()
    agg_p=0.333*p_p+0.333*p_n+0.333*p_l #share coming from homothetic consumption bundle

    pp=push!(pp,p_p)
    pn=push!(pn,p_n)
    pl=push!(pl,p_l)
    agg=push!(agg,agg_p)
end

pricesplot = [pp pn pl agg]
plot(pricesplot, title="Prices dynamics", label=["Primary" "Normal" "Luxury" "Aggregate"], lw=2)
```

```julia
function prices(T,cal)
    #Prices normalized at date t=1
    pp=[100.0]
    pn=[100.0]
    pl=[100.0]
    #Relative prices
    pnp_r=[1.0]
    pnl_r=[1.0]

    for t=1:T-1
        p_p=100*cal.g_primary^(t)+cal.shock_size*randn() # price dynamics of primary goods
        p_n=100*cal.g_normal^(t)+cal.shock_size*randn()# price dynamics of normal goods
        p_l=100*cal.g_luxury^(t)+cal.shock_size*randn() # price dynamics of luxury goods
        
        pnp=p_p/p_n
        pnl=p_l/p_n

        #for matrix
        pp=push!(pp,p_p)
        pn=push!(pn,p_n)
        pl=push!(pl,p_l)

        pnp_r=push!(pnp_r,pnp)
        pnl_r=push!(pnl_r,pnl)
        
    end
    return [pp pn pl pnp_r pnl_r]
end

p=prices(T,cal)

data=DataFrame(p,:auto)

rename!(data,[:"Primary P",:"Normal P", :"LuxuryP",:"Relative P/N",:"Relative L/N"])
```

# Homothetic problem with no income effect

 We start by evaluating our traditional household consumption problem with homothetic preferences. Hence any increase in utility translates into higher consumption in terms of units but same consumption share regarding each product.

 We also assume that prices on each good evolve differently over time so that relative prices are not constant.

```julia
#Expenditure value for all our N individuals for T periods
#individuals are different by their revenue level
H_Exp=[]
for i=1:N
    for t=1:T
    Ex=(u[i]*p[t,1]^(1-cal.σ)+u[i]*p[t,2]^(1-cal.σ)+u[i]*p[t,3]^(1-cal.σ))^(1/(1-cal.σ))

    H_Exp=push!(H_Exp,Ex)
    end
end

H_Exp=reshape(H_Exp,(T,N))
```

```julia
#ROWWISE
#Consumption share over time T for our N individuals
Index=[]
for i=1:N
    for t=1:T
        in=t
        Index=push!(Index,in)
    end
end
Index

H=[]
for i = ["Homothetic"]
    for t=1:T*N
        in=i
        H=push!(H,in)
    end
end
H

Type=[]
for i = ["Poor" "Middle" "Rich"]
    for t=1:T
        in=i
        Type=push!(Type,in)
    end
end
Type

H_share=[]
H_cons=[]

for i=1:N
    for t=1:T
        #SHARE
        ω_p_p=u[i]* (p[t,1]/H_Exp[t,i])^(1-cal.σ)
        ω_n_p=u[i]* (p[t,2]/H_Exp[t,i])^(1-cal.σ)
        ω_l_p=u[i]* (p[t,3]/H_Exp[t,i])^(1-cal.σ)
        H_share=push!(H_share, ω_p_p, ω_n_p, ω_l_p)

        #CONSUMPTION
        c_p=u[i]* (p[t,1]/H_Exp[t,i])^(-cal.σ)
        c_n=u[i]* (p[t,2]/H_Exp[t,i])^(-cal.σ)
        c_l=u[i]* (p[t,3]/H_Exp[t,i])^(-cal.σ)
        H_cons=push!(H_cons, c_p, c_n, c_l)
    end
end
H_share=permutedims(reshape(H_share,N,T*N))
H_cons=permutedims(reshape(H_cons,N,T*N))

H_share_cons=[Index H Type H_share H_cons]

df_H=DataFrame(H_share_cons,:auto)

rename!(df_H,[:"Time %",:"TYPE",:"Household type",:"Primary %", :"Normal %",:"Luxury %",:"Primary Q", :"Middle Q",:"Luxury Q"])
```

```julia
function relativeprice(cal, p_n, T)
    P_pn=cal.A_n/cal.A_p #relative price between primary necessisties and normal good
    P_ln=cal.A_n/cal.A_l #relative price between luxury and normal good
    p_p=P_pn*p_n
    p_l=P_ln*p_n
    return [p_p p_n p_l]
end
p=relativeprice(cal,10)
```

```julia
function utility(u_p,u_m,u_h)
    u_p=u_p
    u_m=u_m
    u_h=u_h
    return [u_p u_m u_h]
end
u=utility(5.0,10.0,30.0)
```

```julia
function expenditure(u)
    E_p=(u[1]^cal.ϵ_p*p[1]^(1-cal.σ)+u[1]^cal.ϵ_n*p[2]^(1-cal.σ)+u[1]^cal.ϵ_l*p[3]^(1-cal.σ))^(1/(1-cal.σ))
    E_m=(u[2]^cal.ϵ_p*p[1]^(1-cal.σ)+u[2]^cal.ϵ_n*p[2]^(1-cal.σ)+u[2]^cal.ϵ_l*p[3]^(1-cal.σ))^(1/(1-cal.σ))
    E_h=(u[3]^cal.ϵ_p*p[1]^(1-cal.σ)+u[3]^cal.ϵ_n*p[2]^(1-cal.σ)+u[3]^cal.ϵ_l*p[3]^(1-cal.σ))^(1/(1-cal.σ))
    return [E_p, E_m, E_h]
end

Exp=expenditure(u)
```

```julia
function trueshare(cal)
    #Poor
    ω_p_p=u[1]^cal.ϵ_p * (p[1]/Exp[1])^(1-cal.σ)
    ω_n_p=u[1]^cal.ϵ_n * (p[2]/Exp[1])^(1-cal.σ)
    ω_l_p=u[1]^cal.ϵ_l * (p[3]/Exp[1])^(1-cal.σ)
    #middle
    ω_p_m=u[2]^cal.ϵ_p * (p[1]/Exp[2])^(1-cal.σ)
    ω_n_m=u[2]^cal.ϵ_n * (p[2]/Exp[2])^(1-cal.σ)
    ω_l_m=u[2]^cal.ϵ_l * (p[3]/Exp[2])^(1-cal.σ)
    #High rich
    ω_p_h=u[3]^cal.ϵ_p * (p[1]/Exp[3])^(1-cal.σ)
    ω_n_h=u[3]^cal.ϵ_n * (p[2]/Exp[3])^(1-cal.σ)
    ω_l_h=u[3]^cal.ϵ_l * (p[3]/Exp[3])^(1-cal.σ)

    return [ω_p_p ω_n_p ω_l_p ω_p_m ω_n_m ω_l_m ω_p_h ω_n_h ω_l_h]
end

share=trueshare(cal)
print(share)
```

### Firm
Each sector $i$ firm produces sectoral output under competition. We adopt a constant-return-to-scale production function with only labor supply and sector specific productivity,
$$Y_i=A_iL_i$$
where $L_i$ denotes the labor supply used in producing output $Y_i$. 
Competitive goods market firm profit maximization imply that wage are same for all sectors and price equal the marginal cost of hiring an additional hour corrected by sector specific productivity. 

It is good to remark that $L_i$ stands for the sum of labor supplied by different household $h$.
So that $L_i = \sum^\mathcal{H}_{h=1}\sum^\mathcal{I}_{i=1}\omega_i^hL^h$ where $L^h$ correspond to the total labor supply regardless of the production sector by household $h$. 
$$\max \limits_{L_i} \Pi = \max p_iY_i-wL_i$$
$$Note: \quad Y_i=\sum^\mathcal{H}_{h=1}C_i^h$$
$$ L_i = \sum^\mathcal{H}_{h=1}\sum^\mathcal{I}_{i=1}\omega_i^hL^h.$$

When we derive with respect to $L_i$, we find $w=p_iA_i$. that can be rewritten as 
$$wL_i=p_iY_i.$$

Good market clearing ensures that $\sum^\mathcal{H}_{h=1}\omega_i^hE^h=p_iY_i$. 

$$wL_i=\sum^\mathcal{H}_{h=1}\omega_i^hE^h.$$
So that at the aggregate level, total expenditure across household and sectors equal total production
$$\sum^\mathcal{I}_{i=1}\sum^\mathcal{H}_{h=1}\omega_i^hE^h=\mathbf{pY}$$
with $\mathbf{p}$ the price level weighted by production size $\mathbf{Y}$.
Regarding relative labor supply across sectors, we see that relative sector size across sectors determine its value.

$$\frac{L_i}{L_j}=\frac{\sum^\mathcal{H}_{h=1}\omega_i^hE^h}{\sum^\mathcal{H}_{h=1}\omega_j^hE^h}$$

Considering that $w=p_iA_i \quad \forall i \in \mathcal{I}$
, then relative price is defined with with relative sector productivity. $$\frac{p_i}{p_j}=\frac{A_j}{A_i}$$
